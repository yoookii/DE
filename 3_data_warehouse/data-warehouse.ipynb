**bigquery-best-practice**
* cost reduction 
    * avoid select * 
    * use clustered or partitioned tables 
    * materialize query results 
* query performance 
    * filter or partitioned columns 
    * denormalizing data 
    * order last
    * place the table with that largest number of rows first, followed by the table with the fewest rows, and then place the remaining tables by decreasing size
